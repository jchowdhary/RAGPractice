### cosine similarity is one of the most common way to find similarity between two vectors.

```python
import numpy as np
def cosine_similarity(vec1,vec2):
    dot_product = np.dot(vec1, vec2)
    norm_a = np.linalg.norm(vec1)
    norm_b = np.linalg.norm(vec2)
    return dot_product / (norm_a * norm_b)

if __name__ == "__main__":
    vec1 = [1, 2, 3]
    vec2 = [4, 5, 6]
    print("Cosine Similarity:", cosine_similarity(vec1, vec2))  # Output: Cosine Similarity: 0.9746318461970762
#!/usr/bin/env python3
```

```python
openai_embeddings = OpenAIEmbeddings(model="text-embedding-ada-002")
openai_vector = openai_embeddings.embed_documents(documents)    
print("OpenAI Document Embeddings:", openai_vector)
print(f'length of documents: {len(openai_vector)}')
print(f'length of first document embedding: {len(openai_vector[0])}')
print(f'first 5 values of first document embedding: {openai_vector[0][:5]}')
print('---------------------------')
print(f'length of second document embedding: {len(openai_vector[1])}')
print(f'first 5 values of second document embedding: {openai_vector[1][:5]}')



```

```python
query = "Lion king is the best movie"
query_vect = openai_embeddings.embed_query(query)
print("Query Embedding:", query_vect)
```

### do the similarity between the query and the documents.

```python
# Do the simialruty between the query and the documents.
for i, doc_vec in enumerate(openai_vector):
    sim = cosine_similarity(query_vect, doc_vec)
    print(f"Cosine Similarity with document {i}: {sim}")
```

```python
# Assume openai_vector contains embeddings for each document
# and query_vect is the embedding for your query

similarities = [cosine_similarity(query_vect, doc_vec) for doc_vec in openai_vector]
best_match_index = similarities.index(max(similarities))
best_match_document = documents[best_match_index]

print(f"Best matching document: {best_match_document}")
print(f"Similarity score: {similarities[best_match_index]}")
```
